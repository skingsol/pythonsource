```python
# 마우스 핸들링

import pyautogui as p



# 마우스 위치 좌표 구하기
position = p.position()
print(position.x, position.y)
```

```python
# moveTo(x좌표, y좌표)
p.moveTo(100, 100, 0.5)
p.moveTo(200, 200, 0.5)
p.moveTo(300, 300, 0.5)
```

```python
# mobe(), moveRel(): 상대 좌표

p.moveTo(300, 300, 0.25)
p.move(100,100, 0.25)
print(p.position())
p.move(200,200,0.25)
```

```python
p.sleep(3)
print(p.position())
```

```python
p.click(2022,32,1)
```

```python
p.sleep(1)
p.click(clicks=2)
```

```python
p.doubleClick(487,142)
```

```python
p.rightClick(500,300)
```

```python
# mouseDown() / mouseUp()

p.sleep(3)
p.moveTo(300,200,0.2)
p.mouseDown()
p.moveTo(500,500,0.2)
p.mouseUp()
```

```python
p.sleep(3)
# 메모장 타이틀 바 위치 알아내기
print(p.position)
```

```python
p.moveTo(359, 149)
p.drag(0,500,duration=0.25)
p.mouseDown()
p.moveTo(359,500,0.2)
p.drag(0,500,duration=0.25)
p.moveTo(359,149,0.2)
p.mouseUp()
```

```python
#mouseInfo()

p.mouseInfo()
```

```python
for i in range(10):
    p.move(100,100)
    p.sleep(1)
```

```python
# 전체 스크린샷
p.screenshot("screen.png")
```

```python
# 특정위치의 스크린 샷
# reginon(x축 상단, y축 상단, width, height)
p.screenshot("region.png", region=(0,0,300,400))
```

```python
# locateOnScreen : 이미지의 화면상 좌표 구하기

screen_locate = p.locateOnScreen("region.png")
print(screen_locate)
```

```python
p.screenshot("region.png", region=(5,205,65,70))
```

```python
file_locate = p.locateOnScreen("region.png", confidence=0.9)
print(file_locate)
p.click(file_locate)
```

```python
file_locate = p.locateOnScreen("checkbox1.png", confidence=0.9)
print(file_locate)
p.click(file_locate)
```

```python
for i in p.locateAllOnScreen("checkbox1.png"):
    print(i)
    p.click(i)

for j in p.locateAllOnScreen("checkoutbox1.png"):
    print(j)
    p.click(j)
```
