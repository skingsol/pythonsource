```python
import requests

s = requests.Session()

# post + data
data = {
    "name":"hong"
}

r = s.post("https://httpbin.org/post",data=data)

print(r.text)

s.close()
```

```python
import requests

s = requests.Session()

# delete + data
data = {
    "name":"hong"
}

r = s.delete("https://httpbin.org/delete",data=data)

print(r.text)

s.close()
```

```python
import requests

s = requests.Session()

# put + data
data = {
    "name":"hong"
}

r = s.put("https://httpbin.org/put",data=data)

print(r.text)

s.close()
```

```python
# get + parameter  ?name=hong&age=14


import requests

s = requests.Session()

param = {"name":"hong","age":14}

r = s.get("https://httpbin.org/get",params=param)

print(r.text)

s.close()
```

```python
import requests

r = requests.get("https://jsonplaceholder.typicode.com/todos/1")
# print(r.text)
print(r.headers)
print(r.encoding)
print(r.json())
print(r.json().keys())
print(r.json().values())
```

```python
r = requests.get("https://jsonplaceholder.typicode.com/users")

for row in r.json():
    for k,v in row.items():
        print(k,":",v)
    print()
```
