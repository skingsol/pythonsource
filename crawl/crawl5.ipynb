```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

def set_chrome_driver():
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--start-maximized")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=chrome_options)
    return driver
```

```python
import time
driver = set_chrome_driver()
driver.get("https://www.daum.net")

print(driver.title)
print(driver.current_url)

print()
print(driver.page_source) # 현재 페이지 소스 가져오기

time.sleep(2)
driver.quit()
```

```python
from selenium.webdriver.common.by import By     #파싱 기준
from selenium.webdriver.common.keys import Keys     #키보드 키 지정

driver = set_chrome_driver()
driver.get("https://www.daum.net")

# 다음 검색어 창에 검색어를 입력한 후 엔터

# 검색창 찾기(파싱) : find_
# driver.find_element(기준, "값")
element = driver.find_element(By.ID,"q")
# element

# 검색어 입력
element.send_keys("아이폰")
# 엔터
element.send_keys(Keys.ENTER)

# 뒤로가기
driver.back()
```

```python
# 다나와
driver = set_chrome_driver()

driver.get("https://www.danawa.com")
#페이지가 로딩될 때까지 잠시 기다리기 : time.sleep(2), driver.implicitly_wait(2)
driver.implicitly_wait(2)
# 로그인 버튼 찾기 =>
login_btn = driver.find_element(By.CLASS_NAME,"btn_user--login")
login_btn.click()  #login_btn.send_keys(Keys.ENTER) 와 동일

driver.implicitly_wait(3)

# 다나와 아이디 입력창 선택
userid = driver.find_element(By.ID,"danawa-member-login-input-id")
# 아이디 입력
userid.send_keys("본인 아이디")

# 다나와 비밀번호 입력창 선택
password = driver.find_element(By.ID,"danawa-member-login-input-pwd")
# 비밀번호 입력
password.send_keys("본인 비밀번호")

# 엔터
password.send_keys(Keys.ENTER)
```

```python
# 한국은행 경제통계 시스템
driver = set_chrome_driver()
driver.get("https://ecos.bok.or.kr")
driver.implicitly_wait(2)
# div.main-left > div.main-icon-menu > ul > li:nth-child(1) > a
# 100 대 통계지표 요소 찾기
menu = driver.find_element(By.CSS_SELECTOR,"div.main-left > div.main-icon-menu > ul > li:nth-child(1) > a").click()
driver.implicitly_wait(2)

# Excel로 다운로드
# div.exelDown.partition-right > button
driver.find_element(By.CSS_SELECTOR,"div.exelDown.partition-right > button").click()
```

```python
# 유튜브 
driver = set_chrome_driver()
driver.get("https://www.youtube.com/")
#driver.implicitly_wait(2)
time.sleep(1)

# 검색창 찾기
search = driver.find_element(By.CSS_SELECTOR,"input#search")

# 검색어 입력
search.send_keys("아이유")
search.send_keys(Keys.ENTER)

#driver.implicitly_wait(2)
time.sleep(3)

# title 출력
titles = driver.find_elements(By.CSS_SELECTOR,"#title-wrapper > h3")
for title in titles:
    print(title.text)
```

```python
# youtube / selenium / BeautifulSoup
from bs4 import BeautifulSoup

driver = set_chrome_driver()
driver.get("https://www.youtube.com/")
time.sleep(1)

# 검색창 찾기
search = driver.find_element(By.CSS_SELECTOR,"input#search")

# 검색어 입력
search.send_keys("아이유")
search.send_keys(Keys.ENTER)
time.sleep(3)

res = BeautifulSoup(driver.page_source)
titles = res.select("#title-wrapper > h3")
for title in titles:
    print(title.text.strip())  #(title.get_text())
```

```python
# iframe
# switch_to.frame() : iframe 안에 있는 요소 찾기 위해서는 iframe 안으로 이동

url = "https://www.w3schools.com/tags/tryit.asp?filename=tryhtml_headers"
driver = set_chrome_driver()
driver.get(url)

# iframe 안으로 들어가기
driver.switch_to.frame("iframeResult")
# h1 태그 찾기
element = driver.find_element(By.TAG_NAME, "h1")
print(element.text)

# iframe 밖으로 나오기
driver.switch_to.default_content()

#//*[@id="textareawrapper"]/div/div[6]/div[1]/div/div/div/div[5]/pre[5]/span/span[4]
element2 = driver.find_element(By.XPATH,"//*[@id='textareawrapper']/div/div[6]/div[1]/div/div/div/div[5]/pre[5]/span/span[4]")
print(element2.text)
```

```python
# 왼쪽 화면에서 h1 태그 안 내용 출력
# 오른쪽 화면에서 HTML 라디오 버튼 클릭

url = "https://www.w3schools.com/tags/tryit.asp?filename=tryhtml5_input_type_radio"
driver = set_chrome_driver()
driver.get(url)

# h1 태그 안 내용 가져와서 출력
element = driver.find_element(By.XPATH, "//*[@id='textareawrapper']/div/div[6]/div[1]/div/div/div/div[5]/pre[5]/span/span[4]")
print(element.text)

# iframe 안으로 들어가기
driver.switch_to.frame("iframeResult")

# 라디오 버튼 클릭하기
driver.find_element(By.ID,"html").click()

# iframe 밖으로 나오기
driver.switch_to.default_content()
```
