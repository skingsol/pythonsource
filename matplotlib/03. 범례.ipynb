## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

### 범례 설정
- https://matplotlib.org/stable/api/legend_api.html

```python
import matplotlib.pyplot as plt
```

```python
# 한글설정 / 마이너스 설정 필수
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
plt.legend()
plt.show()
```

- 두 개의 선을 하나의 차트에 그릴 때

```python
x1 = [1,2,3]
y1 = [1,4,7]
```

```python
plt.plot(x, y, label="data1", color = "blue")
plt.plot(x1, y1, label="data2", color = "red")
plt.legend()
plt.show()
```

```python
plt.plot(x, y, color = "blue")
plt.plot(x1, y1, color = "red")
plt.legend(['data1','data2'])
plt.show()
```

```python
plt.plot(x, y, color = "blue")
plt.plot(x1, y1, color = "red")
plt.legend(['data1','data2'], loc="upper right")
plt.show()
```

```python
plt.plot(x, y, color = "blue")
plt.plot(x1, y1, color = "red")
plt.legend(['data1','data2'], loc=(0.5,0.5)) # x축 , y축 (0~1 사이값사용)
plt.show()
```

><b>2) 범례 폰트 사이즈 지정</b>

```python
plt.plot(x, y, color = "blue")
plt.plot(x1, y1, color = "red")
plt.legend(['data1','data2'], fontsize=16)
plt.show()
```

><b>3) 범례 행 수 지정</b>

```python
plt.plot(x, y, color = "blue")
plt.plot(x1, y1, color = "red")
plt.legend(['data1','data2'], fontsize=16, ncol=2)
plt.show()
```

- [문제] : data1(x : 1, 2, 3; y : 1, 2, 3) 과 data2(x : 1, 2, 3; y : 1, 100, 200) 을 그래프로 출력하시오.

```python
# 범례, red/blue, 제목 : 선차트


plt.plot([1,2,3],[2,4,8], color = "blue")
plt.plot([1,2,3], [1, 100, 200], color = "red")
plt.title("선차트")
plt.legend(["data1","data2"])
plt.show()
```
